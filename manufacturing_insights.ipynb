```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load dataset
df = pd.read_csv('manufacturing_dataset_1000.csv')

print('Dataset Loaded Successfully')
print('Dataset Shape:', df.shape)

# Clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

print('\nAvailable Columns:')
print(df.columns.tolist())

# Explore dataset
print('\nFirst 5 Rows:')
print(df.head())

print('\nDataset Info:')
print(df.info())

print('\nSummary Statistics:')
print(df.describe())

# Histogram distribution
plt.figure(figsize=(10,8))
df.hist(figsize=(10,8))
plt.suptitle('Manufacturing Data Distribution')
plt.show()

# Correlation heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()

# Boxplot for outliers
numeric_cols = df.select_dtypes(include='number').columns
plt.figure(figsize=(10,6))
df[numeric_cols].boxplot()
plt.title('Outlier Detection in Manufacturing Data')
plt.xticks(rotation=45)
plt.show()

# Choose cluster column
cluster_col = 'machine_id'

if cluster_col not in df.columns:
    raise KeyError(f"Column '{cluster_col}' not found in dataset")

clusters = df[cluster_col].unique()
print('\nTotal clusters found:', len(clusters))

# Random cluster sampling
sampled_clusters = pd.Series(clusters).sample(n=min(3,len(clusters)), random_state=42)

print('Selected clusters:', sampled_clusters.tolist())

# Create cluster sample
cluster_sample = df[df[cluster_col].isin(sampled_clusters)]

print('Sample dataset shape:', cluster_sample.shape)

# Save sample dataset
cluster_sample.to_csv('cluster_sample.csv', index=False)

print('Cluster sample saved')

# Original cluster distribution
plt.figure(figsize=(8,5))
df[cluster_col].value_counts().plot(kind='bar', title='Original Cluster Distribution')
plt.xlabel(cluster_col)
plt.ylabel('Count')
plt.show()

# Sample cluster distribution
plt.figure(figsize=(8,5))
cluster_sample[cluster_col].value_counts().plot(kind='bar', title='Sample Cluster Distribution')
plt.xlabel(cluster_col)
plt.ylabel('Count')
plt.show()

# Pie chart distribution
plt.figure(figsize=(6,6))
df[cluster_col].value_counts().plot(kind='pie', autopct='%1.1f%%', title='Cluster Distribution')
plt.ylabel('')
plt.show()

# K-Means clustering
numeric_data = df.select_dtypes(include='number')

scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_data)

kmeans = KMeans(n_clusters=3, random_state=42)
df['kmeans_cluster'] = kmeans.fit_predict(scaled_data)

print('\nK-Means clustering completed')

# Visualize clusters
cols = numeric_data.columns[:2]

plt.figure(figsize=(8,6))
plt.scatter(df[cols[0]], df[cols[1]], c=df['kmeans_cluster'])
plt.xlabel(cols[0])
plt.ylabel(cols[1])
plt.title('K-Means Cluster Visualization')
plt.show()

# Save final dataset
df.to_csv('manufacturing_kmeans_results.csv', index=False)

print('Clustered dataset saved as manufacturing_kmeans_results.csv')

print('Analysis Completed Successfully')
```
